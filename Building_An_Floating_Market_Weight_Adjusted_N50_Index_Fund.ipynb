```
#Downloading the JSON of the datasource of this page - https://www.niftyindices.com/market-data/heat-map-detail?Indexname=NIFTY 50
import pandas as pd
df=pd.read_json('https://iislliveblob.niftyindices.com/jsonfiles/HeatmapDetail/FinalHeatmapNIFTY%2050.json')
df.head()
```

```
df.columns
```

```
Net_Indexmcap_today = df['Indexmcap_today'].sum()
df['IndexWeight']=(df['Indexmcap_today']*100)/(Net_Indexmcap_today)
```

```
MinCapital = max(df['ltP']/df['IndexWeight'])*50
MinCapital
```

```
df['Qty'] = round((max(df['ltP']/df['IndexWeight']))/df['ltP'])
df['Value'] = df['Qty']*df['ltP']
```

```
#The Index Fund
df[['symbol','ltP','IndexWeight','Qty','Value']].sort_values('IndexWeight', ascending=False)
```

```
AllocatedCapital = df['Value'].sum()
AllocatedCapital 
```

```
#Free Float Market Capitalization = Shares outstanding * Price * IWF 
df["free_float_market_cap"] = df['sharesOutstanding']*df['ltP']*df['investableWeightFactor']
net_free_float_market_cap =df["free_float_market_cap"].sum()
net_free_float_market_cap/2060000000000 * 1000
```
